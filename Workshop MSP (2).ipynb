# Dictionary
#### Dictionary is an unordered collection of key:value pairs. It is similar to hashmap.
#### the elements are accessed using the key
#### mutable

```python
dict1={}
print(type(dict1))
```

```python
dict1={'a':'alpha','o':'omega','g':'gamma'}
print(dict1)
dict1['a']

```

### key are immutable,i.e. we cannot change a key. therefore key cannot be a list as those are mutable.
### values are mutable. We can have values of any data type like lists or tuples

```python
dict2={'a':['alpha','apple'],
       'o':['omega','orange'],
       'g':['gamma','guava']}
dict2['a']
dict2['a'][1]
```

```python
dict1['o']="orange"   #mutable
print(dict1)
dict1['b']="beta"   #add values
del dict1['o']      #delete values
print(dict1)
#can also use pop() as in lists
```

## functions

```python
print(dict1.keys())   #print all keys
print(dict1.values())   #print all values
print(dict1.items())    #print all items
```

```python
len(dict2)
```

```python
# clears the whole dictionary
dict2.clear()
dict2
```
